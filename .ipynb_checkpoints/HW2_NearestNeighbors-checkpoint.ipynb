## Question 5. Nearest Neighbors

An implementation of Nearest Neighbor Classifier with $k = 3$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import neighbors

threshold = float(1.0/3.0)
N = 1000
k = 3

def dataGeneration(N,threshold):
    X = np.random.rand(N,2)*10
    y = np.random.rand(N,1)
    for i in range(len(y)):
        if y[i] < threshold:
            y[i] = int(0)
        else: 
            y[i] = int(1)
    return X,y.ravel()

X,y = dataGeneration(N,threshold)

#---------------------------------------------#

X_positive = X[np.where(np.isin(y, 1))[0],:]
X_negative = X[np.where(np.isin(y, 0))[0],:]

print "p: ", threshold
print "Total point count: ", N
print "Possitive point count: ", len(X_positive)
print "Negative point count: ", len(X_negative)

#---------------------------------------------#

def euclideanDistance(instance1, instance2, length):
    distance = 0
    for x in range(length):
        distance += pow((instance1[x] - instance2[x]), 2)
    return math.sqrt(distance)

y_predict = []
for i_0 in range(X.shape[0]):
    distance_array = []
    for i in range(X.shape[0]):
        distance = euclideanDistance(X[i_0],X[i],2)
        distance_array.append(distance)
    k_index = np.argsort(distance_array)[1:k+1]

    possitive_count = 0
    negative_count = 0
    for i in k_index:
        if y[i] == 1:
            possitive_count = possitive_count + 1
        else:
            negative_count = negative_count + 1
    if possitive_count > negative_count:
        predict_label = int(1.0)
    else: 
        predict_label = int(0.0)
    y_predict.append(predict_label)
    
# print(y)
# print(y_predict)    
    
err = sum(pow(y.flatten() - np.array(y_predict),2))/N
print "Prediction error calculated ", err

#---------------------------------------------#

# clf = neighbors.KNeighborsClassifier(k)
# clf.fit(X, y)

# Z = clf.predict(X)

# err_1 = sum(pow(y.ravel() - np.array(Z),2))/N
# print "Prediction error calculated by Sklearn: ", err_1

#---------------------------------------------#

plt.figure(figsize=(10,10)) 
plt.scatter(X_positive[:,0], X_positive[:,1], s=80, c="b", marker="P")
plt.scatter(X_negative[:,0], X_negative[:,1], s=80, c="r", marker="_")
plt.show()
```
